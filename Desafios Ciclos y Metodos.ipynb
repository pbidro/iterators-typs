```python
#banco
global dinero
dinero=0



def consultar_saldo():
    return dinero

def depositar_saldo(saldo, cantidad):
    
    saldo=saldo+cantidad
    global dinero
    dinero=saldo
    return dinero

def girar(saldo,cantidad):
    global dinero
    
    if cantidad<=saldo:
        saldo=saldo-cantidad
        dinero=saldo
        return dinero
    else:
        return False
    
def mostrar_opciones():
    print('1. Consultar saldo')
    print('2. Hacer depósito')
    print('3. Realizar giro')
    print('4. Salir')
    opcion = int(input('Ingrese una opcion:  '))
    return opcion

def mostrar_menu():
    global dinero
    while True:
        
        opcion=mostrar_opciones()
        if opcion==1:
            print('su saldo es '+ str(consultar_saldo()))

        elif opcion==2:
            monto=int(input("Ingrese monto:"))
            depositar_saldo(dinero, monto)
            print('su saldo es '+ str(consultar_saldo()))
        elif opcion==3:
            print('dinero='+str(dinero))
            if dinero is 0:
                print('No puede realizar giros. Su saldo es 0')
            else:
                monto=int(input("Ingrese monto:"))
                if girar(dinero,monto) is not False:
                    print('su saldo es '+ str(consultar_saldo()))
                else:
                    print('No se puede girar esta cantidad. Su saldo es de '+ str(consultar_saldo())+'despues del else')
        elif opcion==4:
            break
        else:
            print('Opcion no valida')
            mostrar_menu()
            
mostrar_menu()
```

```python
#dibujar letra x mejorado

def letra_x_2(num):
    
    #se define caracter para dibujar
    ast='*' 
    
    #se define el centro para los numeros impares
    #la funcion center recibe dos parametros: 1. el largo total de una cadena 2.un caracter
    #el primer parametro hace que se centre el texto en el largo determinado de este por ejemplo:
    # 'hola'.center(6) pasaría a ser ' hola '. el segundo parametro corresponde al caracter con el que se
    #llenarán los espacios vacíos, que por defectio es ' '.
    #por ejemplo 'hola'.center(6,'*') pasaría a ser '*hola*'
    #centrando así una cadena de texto a nuestro antojo
    center = ast.center(num) 
    
    #se define la distancia inicial entre los asteriscos
    distancia = num-2
    
    #se inicializa la lista donde se almacenara la letra
    arreglo_x = [] 
    
    
    
    #en este ciclo se almacentara x y se decrecera en cada iteracion dos unidades correspondiente a los asteriscos
    
    #se iterará la primera mitad del numero ingresado
    for n in range(num//2):
    
    
    #se agrega al arreglo el primer * más los espacios multiplicado por la distancia más el último asterisco
        arreglo_x.append('{0}{1}{0}'.format(ast, ' ' * (distancia)).center(num,' '))
        
    #se disminuye en 2 la distancia entre los asteriscos para restarle los dos asteriscos en cada iteracion
        distancia -= 2 
    
    # descomentar para ver el proceso: 
    #print(arreglo_x)
    
    
    
#la funcion join concatena en texto un elemento iterable
#ejemplo: myTuple = ("John", "Peter", "Vicky")
#         x = "#".join(myTuple)
#         print(x)
#salida: John#Peter#Vicky
#https://www.w3schools.com/python/ref_string_join.asp

   #si el numero es par
    if(not num%2): 
        return '{}\n{}'.format(

            #se concatena el arreglo
            '\n'.join(arreglo_x),

            #se concatena el arreglo recorrido al revés
            '\n'.join(arreglo_x[::-1])
        )
    
    
    #si el numero es impar
    else:
        return '{}\n{}\n{}'.format(
            
        #se concatena el arreglo
        '\n'.join(arreglo_x),
            
        #se concatena el centro
        center,
            
        #se concatena el arreglo recorrido al revés
        '\n'.join(arreglo_x[::-1])
    )
    
#descomentar para probar
#    print(letra_x_2(19))
```
